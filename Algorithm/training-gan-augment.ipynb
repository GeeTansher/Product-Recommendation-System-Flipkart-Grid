```python
from transformers import pipeline, set_seed
import pandas as pd
import numpy as np

# Load the Amazon US reviews beauty dataset
csv_file_path = "/kaggle/input/grid-prs-amazon-dataset/GiftCard.csv"
df = pd.read_csv(csv_file_path)

N = 250  # number of iterations, but there are multiple samples per iteration
product_category = 'Gift Card'
```

```python
def select_short_reviews(dataframe, max_length=300):
    return [idx for idx, text in enumerate(dataframe['review_body']) if len(text) < max_length]
```

```python
selectable_indices = select_short_reviews(df)

generator = pipeline('text-generation', model='gpt2', device=0)  # Use GPT-2. Use GPU. To use CPU, remove the device argument
set_seed(42)  # for reproducibility
```

```python
generated_data = []

for i in range(N):
    print(i)
    indices = np.random.choice(selectable_indices, size=2)

    samples = [df.loc[i] for i in indices]
    in_string = ""
    
    for sample in samples:
        customer_id = sample['customer_id']
        product_id = sample['product_id']
        product_title = sample['product_title']
        rating = sample['rating']
        
        prompt = f"Customer ID: {customer_id}\nProduct ID: {product_id}\nProduct Title: {product_title}\nRating: {rating}\n"
        in_string += prompt
    
    out = generator(in_string, max_length=1024, num_return_sequences=1)
    out_string = out[0]['generated_text']

    start = len(in_string)
    generated_text = out_string[start:].strip()
    
    # Parse the generated text to extract product_id, product_title, and rating
    generated_parts = generated_text.split("\n")  # Adjust the splitting based on your generated content
    if len(generated_parts) >= 4:
        product_id = generated_parts[1].replace("Product ID:", "").strip()
        product_title = generated_parts[2].replace("Product Title:", "").strip()
        rating = generated_parts[3].replace("Rating:", "").strip()
    else:
        product_id = "N/A"
        product_title = "N/A"
        rating = "N/A"
    
    generated_data.append({
        'customer_id': samples[0]['customer_id'],
        'product_id': product_id,
        'product_title': product_title,
        'product_category': product_category,
        'rating': rating,
    })
```

```python
generated_review_date = []

for i in range(N):
    print(i)
    indices = np.random.choice(selectable_indices, size=2)

    samples = [df.loc[i] for i in indices]
    in_string = ""
    
    for sample in samples:
        customer_id = sample['customer_id']
        review_date = sample['review_date']
        
        prompt = f"Customer ID: {customer_id}\nReview Date: {review_date}"
        in_string += prompt
    
    out = generator(in_string, max_length=1024, num_return_sequences=1)
    out_string = out[0]['generated_text']

    start = len(in_string)
    generated_text1 = out_string[start:].strip()
    
    # Parse the generated text to extract product_id, product_title, and rating
    generated_parts1 = generated_text1.split("\n")  # Adjust the splitting based on your generated content
    if len(generated_parts1) >= 2:
        rating = generated_parts1[1].replace("Review Date:", "").strip()
    else:
        review_date = "N/A"
    
    generated_review_date.append({
        'review_date': review_date,
    })
```

```python
# Create a DataFrame from the generated data
generated_df = pd.DataFrame(generated_data)
generated_review_date_df = pd.DataFrame(generated_review_date)
```
